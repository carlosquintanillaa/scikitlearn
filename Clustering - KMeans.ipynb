## Clustering - KMeans

```python
# Load stardard libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load libraries to scale and find clusters
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Read dataset
datos = pd.read_csv("HypotheticalGradesMBA.csv")
```

```python
datos
```

```python
grades = datos.iloc[:,2:8]
```

```python
grades
```

```python
# Create a StandardScaler object
scaler = StandardScaler()
# Fit the scaler to the new DataFrame and transform the dataset
standardized_grades_array = scaler.fit_transform(grades)
# Convert array to DataFrame
stardardized_grades = pd.DataFrame(standardized_grades_array,columns=grades.columns)
```

```python
# Set number of clusters
num_clusters = 2
# Create a KMeans object with the desired number of clusters
cluster_object = KMeans(n_clusters=num_clusters,random_state=42)
# Fit the KMeans algorith to the standardized dataset
cluster_object.fit(stardardized_grades)
```

```python
cluster_labels = cluster_object.labels_
```

```python
cluster_labels
```

```python
cluster_centroids_array = cluster_object.cluster_centers_
```

```python
cluster_centroids_array
```

```python
cluster_centroids_stardardized_units = pd.DataFrame(cluster_centroids_array,columns=grades.columns)
```

```python
cluster_centroids_stardardized_units
```

```python
new_cluster_labels = pd.Series(cluster_object.labels_,name='Group')
grades_augmented = pd.concat([grades,new_cluster_labels],axis=1)
```

```python
grades_augmented
```

```python
cluster_centroids_original_units = grades_augmented.groupby('Group').mean().reset_index()
```

```python
cluster_centroids_original_units
```

```python
cluster_centroids_original_units_melted = cluster_centroids_original_units.melt(id_vars='Group',var_name='Variable',value_name='Mean')
```

```python
cluster_centroids_original_units_melted
```

```python
# Grafico Clusters
plt.figure(figsize=(10,6))
sns.barplot(x='Variable',y='Mean',hue='Group',data=cluster_centroids_original_units_melted)
plt.show()
```
