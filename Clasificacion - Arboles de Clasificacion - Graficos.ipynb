```python
# Load libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
```

```python
# Load data
affaire = pd.read_csv("/Users/carlosq/Desktop/affaire.csv")
```

```python
affaire
```

```python
# Define roles for diamantes1.csv
y = np.where(affaire['affaire']=='si',1,0)
X = affaire.drop(affaire.columns[[6,7]],axis=1)
```

```python
model_tree = DecisionTreeClassifier(max_depth=2)
model_tree.fit(X,y)
```

```python
from io import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(model_tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names=X.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
